```python
#5. Detect and Count coins in the Images
fig, axs = plt.subplots(2,5, figsize=(16,8))
count = 0
for i in os.listdir('./coins'):
  # get the list of images in the particular folder
  train_class = os.listdir(os.path.join('./coins',i))
  # plot 5 images per Folder
  for j in range(5):
        image = cv2.imread('./coins/'+i+'/'+train_class[j])
        greyscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        blurry = cv2.GaussianBlur(greyscale, (17, 17), 0)
        output = cv2.Canny(blurry, 30, 150)
        dilated = cv2.dilate(output, (1,1), iterations = 2)
        (cnts, _) = cv2.findContours(output, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
        cv2.drawContours(image, cnts, -1, (0, 255, 0), 2)
        print("I found %i coins" % len(cnts))
        axs[count][j].imshow(image)
        axs[count][j].set_title(i, fontsize=30)
  count+=1
fig.tight_layout()
```
