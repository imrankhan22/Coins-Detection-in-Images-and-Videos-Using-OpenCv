```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('Video.mp4')
if (cap.isOpened()== False):
    print("Error opening video file")
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(grey, (17, 17), 0)
        outline = cv2.Canny(blurred, 30, 150)
        dilated = cv2.dilate(outline, (1,1), iterations = 2)
        (cnts, _) = cv2.findContours(outline, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
        cv2.drawContours(frame, cnts, -1, (0, 255, 0), 2)
        print("I found %i coins" % len(cnts))
        cv2.imshow('Frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
